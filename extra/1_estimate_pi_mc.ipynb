```python
# Simple Monte Carlo - Estimate Pi
import random

sample_times = 900000
sample_cnt = 0
in_circ_cnt = 0

# simulation
for i in range(sample_times):
    random_x = random.random() * 2.0 - 1
    random_y = random.random() * 2.0 - 1
    
    # check if in the circle
    if random_x**2 + random_y**2 <= 1:
        in_circ_cnt += 1
    sample_cnt += 1

# pi/4
print(in_circ_cnt/sample_cnt)
# pi
print(4 * (in_circ_cnt/sample_cnt))
```

```python
# Metropolis Monte Carlo - Estimate Pi
sample_times = 900000
sample_cnt = 0
in_circ_cnt = 0
step_size = 2.0
accept_cnt = 0

old_x = random.random() * 2.0 - 1
old_y = random.random() * 2.0 - 1

# simulation
for i in range(sample_times):
    new_x = old_x + step_size * (2*random.random()-1)
    new_y = old_y + step_size * (2*random.random()-1)
    
    # ensure its in the square
    if abs(new_x) <= 1 and abs(new_y) <= 1:
        old_x = new_x
        old_y = new_y
        accept_cnt += 1
    
    # check if in the circle
    if old_x**2 + old_y**2 <= 1:
        in_circ_cnt += 1
    sample_cnt += 1

# pi/4
print(in_circ_cnt/sample_cnt)
# pi
print(4 * (in_circ_cnt/sample_cnt))
# acceptance rate # larger step -> lower acceptance (higher accuracy when acceptance ~ 0.25)
print(accept_cnt/sample_cnt)
```
